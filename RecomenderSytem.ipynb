```python
# Extracting text  from the genres column
import ast
def convert_genres(text):
    genres = []
    try:
        for i in ast.literal_eval(text):
            genres.append(i['name'])
    except:
        pass
    return genres
content_data['genres'] = content_data['genres'].apply(convert_genres)
content_data['genres'].head()
```

```python
content_data.iloc[0].keywords
```

```python
# extracting text from the keywords column
def extract_keywords(text):
    keywords = []
    for item in ast.literal_eval(text):
        keywords.append(item['name'])
    return keywords
content_data['keywords'] = content_data['keywords'].apply(extract_keywords)
```

```python
content_data.iloc[0].production_companies
```

```python
# Extracting text from the production_companies column
def extract_production_companies(text):
    companies = []
    for item in ast.literal_eval(text):
        companies.append(item['name'])
    return companies
content_data['production_companies'] = content_data['production_companies'].apply(extract_production_companies)
```

```python
content_data.iloc[0].production_countries
```

```python
# extracting text from the production_companies column
def extract_production_countries(text):
    countries = []
    for item in ast.literal_eval(text):
        countries.append(item['name'])
    return countries
content_data['production_countries'] = content_data['production_countries'].apply(extract_production_countries)
```

```python
content_data.iloc[0].spoken_languages
```

```python
# extracting text from the spoken_languages column
def extract_spoken_languages(text):
    languages = []
    for item in ast.literal_eval(text):
        languages.append(item['name'])
    return languages
content_data['spoken_languages'] = content_data['spoken_languages'].apply(extract_spoken_languages)
```

```python
# Converting list cols to string
list_cols = [
    'genres',
    'keywords',
    'production_companies',
    'production_countries',
    'spoken_languages'
]

for col in list_cols:
    content_data[col] = content_data[col].apply(
        lambda x: " ".join([i.replace(" ", "") for i in x]) if isinstance(x, list) else str(x)
    )

# Ensuring that cols are string
text_cols = ['overview', 'tagline']

for col in text_cols:
    content_data[col] = content_data[col].apply(str)
```

```python
# Combing all text data into a single column
content_data['tags'] = (
    content_data['genres'] + " " +
    content_data['overview'] + " " +
    content_data['keywords'] + " " +
    content_data['tagline'] + " " +
    content_data['production_companies'] + " " +
    content_data['production_countries'] + " " +
    content_data['spoken_languages']
)
```

```python
content_data = content_data[['title', 'tags']]
```

```python
# creating idx for movie titles
content_data = content_data.reset_index(drop=True)

idx = pd.Series(content_data.index, index=content_data['title']).drop_duplicates()

content_data['movie_index'] = content_data.index
```

# Natural Language Processing

```python
import re
import string
def clean_text(text):
    text = str(text)               # ensure string
    text = text.lower()           # lowercase
    text = re.sub(r'\d+', '', text)  # remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # punctuation removal
    text = " ".join(text.split())  # extra spaces removal
    return text
content_data['tags'] = content_data['tags'].apply(clean_text)
```

```python
# Lower case the movie titles
def lower_case(text):
    return text.lower()
content_data['title'] = content_data['title'].apply(lower_case)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_text(text):
    words = text.split()
    return " ".join([ps.stem(word) for word in words])
content_data['tags'] = content_data['tags'].apply(stem_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = vectorizer.fit_transform(content_data['tags']).toarray()
```

```python
print(vectors.shape)
print(vectorizer.get_feature_names_out())
print(vectorizer.get_metadata_routing())
print(vectorizer.get_stop_words())
print(vectorizer.get_params())
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend_df(movie, top_n=5):

    # normalize input
    movie = movie.strip().lower()

    # create lowercase mapping of idx
    idx_lower = {k.lower(): v for k, v in idx.items()}

    # check existence
    if movie not in idx_lower:
        return "Movie not found in dataset"

    movie_index = idx_lower[movie]
    distances = similarity[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:top_n+1]

    data = []

    for i, score in movie_list:
        data.append({
            "Movie": content_data.iloc[i].title,
            "Similarity": round(float(score), 3)
        })

    return pd.DataFrame(data)
```

```python
result = recommend_df("pirates of the Caribbean: at world's end")
print(result)
```
